
# 🔍 Recherche et sélection d’une plateforme IA Cloud
## Module AIC-108 – Optimisation intelligente du trafic d’ascenseurs

---

## 🎯 Objectif du document

Ce document vise à identifier la plateforme d’intelligence artificielle la plus appropriée pour résoudre le problème suivant :

> **Rocket Elevators** fait face à des temps d’attente élevés pour ses ascenseurs, en particulier aux heures de pointe, dans des immeubles de 50+ étages.

Nous souhaitons concevoir une solution IA capable de prédire la demande, guider les ascenseurs de manière optimale, et automatiser le tout dans un environnement cloud fiable et réactif.

---

## ❓ Problématique technique à résoudre

> **Comment peut-on utiliser des outils d’intelligence artificielle, combinés à un système d’automatisation cloud, pour optimiser en temps réel la gestion du trafic d’ascenseurs dans un immeuble de 50+ étages ?**

---

## 🔧 Critères de sélection d’un outil IA

Pour répondre à ce besoin, la plateforme IA idéale devra offrir :
- 🔄 Traitement en **temps réel**
- 📊 Capacités de **prédiction** / classification
- ☁️ **Intégration Cloud** (automatisation, monitoring)
- 🛠️ Documentation solide et simplicité d’usage
- 💸 Option gratuite ou low-cost
- 🔁 Compatibilité CI/CD (pour mises à jour continues)

## ✅ Outil recommandé

Après avoir analysé les principales plateformes, nous recommandons d’utiliser **Azure Machine Learning Studio**, pour les raisons suivantes :
- Interface simple et pédagogique, parfaite pour un projet d’école
- Intégration fluide avec les services cloud et pipelines CI/CD
- Prise en charge du **AutoML**, idéal pour tester rapidement des modèles
- Gratuit jusqu’à un certain usage (bon point pour l’expérimentation)
- Documentation claire, bon support communautaire

---

## ⚠️ Contraintes / Inconvénients rencontrés

- Certaines plateformes comme SageMaker sont très puissantes mais demandent des compétences avancées en DevOps
- L’accès à des environnements en temps réel peut être limité dans les versions gratuites
- La mise en place d’un pipeline complet (CI/CD, monitoring, déploiement) demande plusieurs outils à intégrer

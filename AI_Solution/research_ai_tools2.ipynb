
## 🧠 Comment chaque plateforme IA répond au problème d'optimisation du trafic d’ascenseurs

---

### 1. ☁️ AWS SageMaker

AWS SageMaker est une plateforme puissante pour entraîner et déployer des modèles prédictifs, particulièrement dans des environnements de production à grande échelle.

**Comment SageMaker répond au problème :**
- ✅ Peut entraîner des modèles supervisés pour prédire les moments/étages de forte demande.
- ✅ Déploiement en tant qu’API REST via des endpoints gérés.
- ✅ Intégration directe avec AWS Lambda pour une réponse automatisée en temps réel.
- ✅ Monitoring avec CloudWatch pour relancer automatiquement l'entraînement.
- ⚠️ Plus complexe à prendre en main et moins adapté à un environnement académique.

---

### 2. ☁️ Azure Machine Learning Studio

Azure ML Studio est une plateforme cloud no-code/low-code idéale pour expérimenter et automatiser un modèle IA complet, de l’entraînement au déploiement.

**Comment Azure ML répond au problème :**
- ✅ AutoML intégré pour prédire les pics d’utilisation d’ascenseurs.
- ✅ Déploiement rapide en endpoint REST utilisable en temps réel par un logiciel d’ascenseur.
- ✅ Intégration avec Azure DevOps pour automatiser tout le pipeline (training + deployment).
- ✅ Azure Monitor permet de surveiller les performances du modèle et relancer un entraînement si nécessaire.
- ✅ Interface pédagogique et bien documentée : idéale pour un projet d’école.

---

### 3. ☁️ Google Vertex AI

Vertex AI est une plateforme unifiée de Google Cloud qui permet de construire des systèmes IA industriels et scalables, avec une grande flexibilité.

**Comment Vertex AI répond au problème :**
- ✅ Possibilité de prédire les futurs appels d’ascenseurs avec AutoML ou training manuel.
- ✅ Déploiement rapide d’un modèle sous forme d’endpoint REST ou API Gateway.
- ✅ Intégration forte avec BigQuery pour l’analyse des données historiques.
- ✅ Bon support CI/CD avec Cloud Build et GitHub Actions.
- ⚠️ Environnement GCP plus difficile d’accès pour un public non expérimenté.

---

### 4. ☁️ IBM Watson Studio

Watson Studio est une plateforme accessible pour les débutants qui souhaitent expérimenter avec des modèles prédictifs en environnement cloud.

**Comment IBM Watson répond au problème :**
- ✅ Peut construire des modèles de prédiction via AutoAI (AutoML) pour anticiper les appels d’ascenseur.
- ✅ API déployable avec REST pour être interrogée en temps réel.
- ✅ Bon environnement pour le prototypage rapide avec interface no-code.
- ⚠️ Intégration DevOps moins poussée que les autres plateformes modernes.
- ⚠️ Moins orienté vers l’automatisation complète et continue.

---

## ✅ Conclusion générale

Chaque plateforme offre une capacité à résoudre le problème, mais **Azure ML Studio** offre l'équilibre parfait entre :
- 📊 Prédiction fiable
- 🚀 Déploiement rapide
- 🔁 Automatisation continue
- 🧠 Simplicité d’apprentissage

C’est donc la meilleure solution pour répondre aux besoins de Rocket Elevators.
